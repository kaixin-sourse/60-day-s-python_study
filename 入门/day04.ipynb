```python
# 使用众数填充缺失值
import pandas as pd
data = pd.read_csv('data.csv') # 重新读一遍数据
mode = data['Annual Income'].mode()
mode
```

```python
mode = mode[0]
```

```python
#众数填补
data['Annual Income'].fillna(mode,inplace=True)
# 检查下是否有缺失值 # 查一下缺失值个数
data['Annual Income'].isnull().sum()
```

```python
data.columns
```

```python
import numpy as np
a = np.array([1,2,3]) # 一维数组
a.tolist() # 转换为列表
```

```python
c = data.columns.tolist()
type(c)
```

```python
c
```

```python
type(c[0])
```

```python
# 循环遍历c这个列表中的每一列
for i in c:
    if data[i].dtype != 'object':
        if data[i].isnull().sum() > 0:
            mean_value = data[i].mean()
            #用均值填充缺失值
            data[i].fillna(mean_value, inplace=True)
data.isnull().sum()
```
